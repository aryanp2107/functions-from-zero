####Most useful functions

```python
def my_worker(fruit):
  statement = f"My favorite meal is chicken and {fruit}"
  meals = {"dinner": statement}
  return meals
```

```python
meal = my_worker("apple")
meal
```

```python
type(meal)
```

```python
meal
```

```python
meal.keys()
```

```python
meal.values()
```

```python
def human(food):
  consume = food["dinner"]
  return consume
```

```python
my_meal = human(meal)
print(f"I want my meal. It will be: {my_meal}")
```
